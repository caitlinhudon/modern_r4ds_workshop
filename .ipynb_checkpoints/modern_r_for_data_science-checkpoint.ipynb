## Sampling

Use cross-fold to check our feature engineering steps (if we're transforming, is it stable when sampling?)

```r
titanic_smp = initial_split(titanic_df, prop = .7, strata = 'survived')

titanic_train <- training(titanic_smp)
titanic_test <- testing(titanic_smp)
```

```r
titanic_train_cv <- vfold_cv(titanic_train)
```

Create cross-validation samples, create a GLM for each sample, combine them, then plot!

```r
titanic_train_cv %>%
    # each one of these is a split we're doing -- pluck grabs a column 
    pluck("splits") %>%
    # do transformations to make this ten training samples of data
    map(as_data_frame)%>%
    # map a very simple model (who survived by class) -- will do 10 models
    map(~glm(survived~pclass, data = .)) %>%
    # combines all of the dataframes into one big dataframe
    map_df(broom::tidy, .id = "cv") %>%
    ggplot() + 
    aes(x = estimate) +
    geom_density() + 
    facet_wrap(~term, scale = "free")
```

Do these tend towards a normal distribution? Are there problematic records that seem to cause the GLM to be bad?

## Feature Engineering

* Recipe: transform an input into an output given a number of pre-defined steps (bake our data!)
* We end up with a model object that we can apply to anything that comes our way
* Encapsulating feature transformations into a model means that basically two lines need to be put into prod:
    * Line 1: Bake the data (recipe object)
    * Line 2: Apply the model (model object)

```r
# recipe needs two things -- a dataset (to get schema), and a formula (outcome variable by all other variables)
# here, survived will get a tag of outcome, and everything else will get a tag of predictor
# can also add columns with custom roles (like ID) -- handy because data in prod will have foreign keys, IDs, etc
# this is a full pipeline
# note, you can also create your own custom steps

recipe(titanic_train, survived ~ .,) %>%
    step_scale(all_numeric()) %>%
    # columns we want to drop -- here, body ID
    step_rm(body) %>%
    prep(titanic_train) ->
    # pipe it into fe_titanic (!)
    fe_titanic
```

```r
fe_titanic
```

```r
titanic_train_cleaned = bake(fe_titanic, titanic_train)
# scaled out outcome column (maybe not desirable), but also scaled others
# and body has been removed (as intended)
# might need to add some NLP -- tidytext is in tidymodels
```

## Build Models!

```r
saveRDS(fe_titanic, "fe_titanic.rds")
```

```r
# creates a series of models
# furrr will train models on multiple CPUs, just change map to future_map!
# models end up in a list -- but we need to be able to evaluate them!
list(all = as.factor(survived) ~ sex + pclass + age ,
    women_n_chillun = as.factor(survived) ~ sex + age) %>%
    map(~glm(formula = ., data = titanic_train_cleaned, family = "binomial")) ->
    candidate_models
```

```r
# don't try this at home -- 
# adding model of a different type to look at how it impacts 

candidate_models[["lm"]] = lm(survived ~ sex + pclass, data = titanic_train_cleaned)
```

```r
# check out variables in the model with coefficients
candidate_models %>%
    map_df(tidy, .id = "model")
```

```r
# check out the model fit statistics
candidate_models %>%
    map_df(glance, .id = "model")
```

```r
# get fitted data with predictions for all of our models
# can build predicted vs. actual charts at the point level 
# to compare models against one another (on training data)
candidate_models %>%
    map_df(augment, .id = "model")
```

You can have a list of a bunch of different models, construct this kind of thing to compare a bunch of different model types' performance on the same dataset

```r
# let's see how this looks on our test data
titanic_test %>%
    bake(fe_titanic, .) %>%
    add_predictions(candidate_models$all, var = "all")
```

```r
# ... and a more cleaned up version -- adds a new column to our three datasets
# now we have three copies of the same data, so try to get to next stage
#
candidate_models %>%
    map_df(~add_predictions(bake(fe_titanic, titanic_test),.), .id = "model") %>%
    spread(model, pred)
```

```r
candidate_models %>%
    map_df(~add_predictions(bake(fe_titanic, titanic_test), 
                           .), .id = "model") %>%
    spread(model, pred)
```

## Model Evaluation

```r
# confusion matrix -- can start to stack these by model or by cut-off point threshhold

titanic_test %>%
    bake(fe_titanic, .) %>%
    add_predictions(candidate_models$all, var = "all") %>%
    mutate(survived = as.factor(survived == 0)) %>%
    mutate(all_class = as.factor(all < 0)) %>%
    yardstick::conf_mat(survived, all_class) %>%
    pluck("table") %>%
    as.data.frame()
```
